#**2.1.1 HEIGHT AND WEIGHT**

```python
fd['Height'].unique()         
```

The height column contains inconsistent measurement units, with some values expressed in centimeters (cm) and others in feet/inches. To address this issue, we will convert all values to centimeters. This involves converting all feet/inches measurements to centimeters, and leaving the centimeter measurements unchanged. By doing so, we will have a consistent unit of measurement for the entire column.

```python
#Converting Height column to cm
def convert_height(height_str):
  if "'" in height_str and '"' in height_str:
    #converting feet, inches to cm
    feet, inches = height_str.split("'")
    total_inches = int(feet)* 12 + int(inches[:-1])
    height_cm = total_inches * 2.54
  elif 'cm' in height_str:
    #converting cm to feet, inches and back to cm
    height_cm = int(height_str[:-2])
    total_inches = round(height_cm / 2.54)
  else:    
   # Assume the height is already in feet/inches
        feet, inches = height_str.split("'")
        total_inches = int(feet) * 12 + int(inches[:-1])
        height_cm = total_inches * 2.54
  return round(height_cm)  #rounding the figures to the nearest whole number 

fd['Height'] = fd['Height'].apply(convert_height)

fd['Height'].unique()   
```

```python
fd= fd.rename(columns={'Height':'Height(cm)'})

fd.columns
```

```python
fd['Weight'].unique()
```

We also noticed that the weight column had inconsistencies in measurement units, as some values were recorded in pounds (lbs) while others were in kilograms (kg). To ensure consistency, we decided to convert all weights recorded in lbs to kg.

```python
#converting weight to all kg

def convert_weight(weight_con):
  if "kg" in weight_con:
    #converting kg to kg
    weight_kg = float(weight_con.replace("kg", "").strip())
  elif "lbs" in weight_con:
    #converting lbs to kg 
    weight_lbs = float(weight_con.replace("lbs","").strip())
    weight_kg = weight_lbs/2.20462
  else:
    #incase it doesnt have either
    weight_kg = float(weight_con.strip()) 

  return(round(weight_kg))  

fd['Weight']= fd['Weight'].apply(convert_weight)

fd['Weight'].unique()
```

```python
fd= fd.rename(columns={'Weight':'Weight(kg)'})

fd.columns
```

We observed earlier that the columns "Value, Wage and Release Clause" have incorrect data types. We will correct these as well as replace "M" with millions and "K" with thousands numerically to ensure consistency in the column.


```python
#converting to a single currency 
def convert_value(value_str):
    value_str = value_str.replace('€', '')
    if 'M' in value_str:
      #converting M to millions
        value_str = float(value_str.replace('M', '')) * 1000000
    elif 'K' in value_str:
      #converting K to thousands 
        value_str = float(value_str.replace('K', '')) * 1000
    else:
        value_str = float(value_str)
    return value_str
fd['Value']= fd['Value'].apply(convert_value)

fd['Value'].unique()
```

```python
fd= fd.rename(columns={'Value':'Value(€)'})
```

**CONVERTING WAGE**

```python
def convert_wage(wage_zz):
    wage_zz = wage_zz.replace('€', '')
    if 'M' in wage_zz:
        wage_zz = float(wage_zz.replace('M', '')) * 1000000
    elif 'K' in wage_zz:
        wage_zz = float(wage_zz.replace('K', '')) * 1000
    else:
        wage_zz = float(wage_zz)
    return wage_zz
fd['Wage']= fd['Wage'].apply(convert_wage)

fd['Wage'].unique()
```

After examining the data further, it was apparent that the Contract column contained inconsistencies because some players were on loan. To address this, I split the Contract column into two separate columns: "Contract Start Year" and "Contract End Year." I used regex in the python library to extract the regular date pattern and create a new column called Loan Date Start to distinguish players on loan from those with full-term contracts.

```python
fd['Contract'] = fd['Contract'].str.replace('~', "-")

fd['Contract'].unique()
```

```python
fd[['Contract Start Year', 'Contract End Year']] = fd['Contract'].str.split(' - | On Loan', expand=True)

fd.columns
```

```python
fd['Contract Start Year'].unique()
```

```python
fd['Contract End Year'].unique()
```

```python
import re

# defining the regex pattern
pattern = r'(\w{3}\s\d{1,2},\s\d{4})'

# extract the year from rows that match the pattern
fd['Loan Date Start'] = fd['Contract Start Year'].str.extract(pattern)

# checking the new column
fd['Loan Date Start'].unique()
```

```python
fd['Contract Start Year'].unique()
```

```python
# defining the regex pattern
pattern = r'[A-Za-z]{3}\s\d{1,2},\s\d{4}'

# creating a mask for the rows that match the pattern
mask = fd['Contract Start Year'].str.contains(pattern)

# replacing the matching values with null
fd.loc[mask, 'Contract Start Year'] = np.nan

# checking for unique values
fd['Contract Start Year'].unique()
```

```python
fd['Contract Start Year'].isnull().sum()
```

```python
fd['Contract End Year']= fd['Contract End Year'].replace('', pd.NA)

fd['Contract End Year'].isna().sum()
```
